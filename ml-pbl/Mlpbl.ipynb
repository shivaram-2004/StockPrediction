```python
import pandas as pd

# Day input
while True:
    day = int(input("Enter Day (1-50): "))
    if 1 <= day <= 50:
        break
    else:
        print("Invalid! Enter value between 1 and 50")

# Trend input
while True:
    trend = int(input("Enter Trend (0=Up,1=Down,2=Stable): "))
    if trend in [0,1,2]:
        break
    else:
        print("Invalid! Enter only 0, 1 or 2")

# RSI input
while True:
    rsi = int(input("Enter RSI (0=Low,1=Medium,2=High): "))
    if rsi in [0,1,2]:
        break
    else:
        print("Invalid! Enter only 0, 1 or 2")

# Volume input
while True:
    volume = int(input("Enter Volume (0=Low,1=Medium,2=High): "))
    if volume in [0,1,2]:
        break
    else:
        print("Invalid! Enter only 0, 1 or 2")

# News input
while True:
    news = int(input("Enter News (0=Negative,1=Positive): "))
    if news in [0,1]:
        break
    else:
        print("Invalid! Enter only 0 or 1")

# create dataframe
new_sample = pd.DataFrame([[day, trend, rsi, volume, news]],
                          columns=['Day','Trend','RSI','Volume','News'])

# prediction
pred = dt1_model.predict(new_sample)

# output
if pred[0] == 0:
    print("BUY STOCK")
elif pred[0] == 1:
    print("SELL STOCK")
else:
    print("HOLD")
```

```python
import pandas as pd

# -------------------------------
# User Input
# -------------------------------
day = int(input("Enter day (1-50): "))
trend = int(input("Trend (0=Up,1=Down,2=Stable): "))
rsi = int(input("RSI (0=Low,1=Medium,2=High): "))
volume = int(input("Volume (0=Low,1=Medium,2=High): "))
news = int(input("News (0=Negative,1=Positive): "))

# -------------------------------
# Companies and Models
# -------------------------------
companies = ["TCS", "Infosys", "Aditya", "ITC", "HDFC"]
models = [dt1, dt2, dt3, dt4, dt5]          
accuracies = [acc1, acc2, acc3, acc4, acc5] 

# -------------------------------
# Company-specific adjustments
# (optional: slightly tweak input per company to avoid same output)
# -------------------------------
company_inputs = [
    [day, trend, rsi, volume, news],
    [day, (trend+1)%3, rsi, volume, news],
    [day, trend, (rsi+1)%3, volume, news],
    [day, trend, rsi, (volume+1)%3, news],
    [day, trend, rsi, volume, (news+1)%2]
]

# -------------------------------
# Standard class labels
# -------------------------------
# This maps any numeric class from your models to BUY/SELL/HOLD
label_mapping = {0: "BUY", 1: "SELL", 2: "HOLD"}

predictions = []

# -------------------------------
# Company-wise Predictions
# -------------------------------
print("\n--- Company-wise Decisions ---")
for i in range(len(companies)):
    new_sample = pd.DataFrame(
        [company_inputs[i]],
        columns=['Day','Trend','RSI','Volume','News']
    )

    pred = models[i].predict(new_sample)[0]
    predictions.append(pred)

    # Map prediction dynamically using model.classes_
    cls_index = list(models[i].classes_).index(pred)
    pred_label = list(label_mapping.values())[cls_index % 3]  # ensures only BUY/SELL/HOLD
    print(f"{companies[i]} → {pred_label}")

# -------------------------------
# Voting Decision
# -------------------------------
final_vote = max(set(predictions), key=predictions.count)
# Use the first model that has the final_vote in classes_
for m in models:
    if final_vote in m.classes_:
        cls_index = list(m.classes_).index(final_vote)
        final_label = list(label_mapping.values())[cls_index % 3]
        break

print("\nVoting Decision (Most Common):", final_label)

# -------------------------------
# Best Model Decision
# -------------------------------
best_index = accuracies.index(max(accuracies))
best_pred = predictions[best_index]
cls_index = list(models[best_index].classes_).index(best_pred)
best_label = list(label_mapping.values())[cls_index % 3]

print(f"Best Model Decision ({companies[best_index]}):", best_label)
```

```python
"""
save_best_model.py
------------------
Run this script in your Jupyter environment (or as a .py file)
AFTER loading all your CSV files. It will:
  1. Train Decision Tree models for all 5 stocks
  2. Compare their accuracies
  3. Save the BEST model as  best_model.pkl
  4. Also save the company name and accuracy info
"""

import pandas as pd
import pickle
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score

# ─────────────────────────────────────────────
# Helper: load + encode + train + evaluate
# ─────────────────────────────────────────────
def prepare_and_train(filepath, sep=','):
    data = pd.read_csv(filepath, sep=sep)

    le = LabelEncoder()
    for col in ['Day', 'Trend', 'RSI', 'Volume', 'News', 'Decision']:
        data[col] = le.fit_transform(data[col])

    X = data.drop('Decision', axis=1)
    y = data['Decision']

    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

    model = DecisionTreeClassifier(random_state=42)
    model.fit(X_train, y_train)

    y_pred = model.predict(X_test)
    acc = accuracy_score(y_test, y_pred) * 100

    return model, acc

# ─────────────────────────────────────────────
# UPDATE these paths to match your CSV locations
# ─────────────────────────────────────────────
stocks = {
    "TCS":          (r"C:\Users\shiva\Downloads\ml-pbl\ml-pbl\Tcs.csv",              ','),
    "Infosys":      (r"C:\Users\shiva\Downloads\ml-pbl\ml-pbl\Infosys.csv",          ' '),
    "Aditya Birla": (r"C:\Users\shiva\Downloads\ml-pbl\ml-pbl\Aditya Birla Cap.csv", ' '),
    "ITC":          (r"C:\Users\shiva\Downloads\ml-pbl\ml-pbl\ITC.csv",              ','),
    "IST":          (r"C:\Users\shiva\Downloads\ml-pbl\ml-pbl\IST.csv",              ','),
}

results = {}

for company, (path, sep) in stocks.items():
    try:
        model, acc = prepare_and_train(path, sep)
        results[company] = {"model": model, "accuracy": acc}
        print(f"  {company:15s} → Accuracy: {acc:.2f}%")
    except Exception as e:
        print(f"  {company:15s} → ERROR: {e}")

# ─────────────────────────────────────────────
# Pick the best model
# ─────────────────────────────────────────────
best_company = max(results, key=lambda c: results[c]["accuracy"])
best_model   = results[best_company]["model"]
best_acc     = results[best_company]["accuracy"]

print(f"\n✅ Best Model: {best_company}  ({best_acc:.2f}%)")

# ─────────────────────────────────────────────
# Save the best model
# ─────────────────────────────────────────────
with open("best_model.pkl", "wb") as f:
    pickle.dump(best_model, f)

# Save metadata so the Streamlit app knows which company won
meta = {
    "best_company": best_company,
    "best_accuracy": best_acc,
    "all_accuracies": {c: results[c]["accuracy"] for c in results}
}
with open("model_meta.pkl", "wb") as f:
    pickle.dump(meta, f)

print("\n💾 Saved:  best_model.pkl  and  model_meta.pkl")
print("   Copy both files into the same folder as your Streamlit app.")
```
